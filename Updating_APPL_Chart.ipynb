### The purpose of this notebook is to get updates from yahoo finance daily to not download a csv file everyday

```python
page = requests.get('https://finance.yahoo.com/quote/AAPL/history?period1=1565481600&period2=1597276800&interval=1d&filter=history&frequency=1d')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.prettify()
```

```python
warning = soup.find('table', class_ = 'W(100%) M(0)')
warning
```

```python
columns = warning.find_all('td')
columns[0:7]
```
